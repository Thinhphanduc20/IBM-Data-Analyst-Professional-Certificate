## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
#import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_column',None)
pd.set_option('display.max_row',100)
pd.set_option('display.width',1200)
#Run SQL
df = pd.read_sql_query("SELECT * FROM master",conn)
df.head()
np.random.seed(42)
x = np.random.normal(size=1000)

plt.figure(figsize=(12,6))
sns.distplot(a=df["ConvertedComp"],bins=20,kde= False)
plt.show()

Query2 = """
SELECT ConvertedComp 
FROM master
"""

ConvertedComp= pd.read_sql_query(Query2, conn)
ConvertedComp.plot(kind='hist', figsize=(10, 8))
```

Plot a box plot of `Age.`

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Cach dai
df.WorkWeekHrs.median()
df.CodeRevHrs.median()
age = df[(df["Age"] >= 30.0) & (df["Age"] <= 35.0)]
age
plt.figure(figsize=(10,10))
plt.stackplot(age.WorkWeekHrs, age.CodeRevHrs)
plt.show()

# Cach SQL
Query8 = """
SELECT avg(WorkWeekHrs), avg(CodeRevHrs), Age 
FROM master
WHERE AGE between 30 and 35
GROUP BY AGE
"""
stacked_chart= pd.read_sql_query(Query8, conn)
WorkWeekHrs=stacked_chart['avg(WorkWeekHrs)']
CodeRevHrs = stacked_chart['avg(CodeRevHrs)']
Age=stacked_chart['Age']
fig, ax = plt.subplots()
ax.bar(Age,WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age,CodeRevHrs, bottom=WorkWeekHrs, label='CodeRevHrs')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
# Cach dai
age2 = df[(df["Age"] >= 45.0) & (df["Age"] <= 60.0)]
age2.head()
age2.ConvertedComp
age2.ConvertedComp.plot(figsize=(15,10))
plt.show()
# Cach SQL
Query9 = """
SELECT avg(ConvertedComp), Age 
FROM master
WHERE AGE between 45 and 60
GROUP BY AGE
"""
line_chart= pd.read_sql_query(Query9, conn)
line_chart
plt.plot(line_chart['Age']),line_chart['avg(ConvertedComp)']
print(line_chart)
```
